# Data Preprocessing

This notebook demonstrates the text cleaning and preprocessing pipeline for the email spam dataset.

```python
import pandas as pd
from src.preprocess import clean_text

# Load data
emails = pd.read_csv('../data/emails.csv')

# Clean a sample email
sample = emails['text'].iloc[0]
print('Original:', sample[:200])
print('Cleaned:', clean_text(sample))
```

```python
# Apply cleaning to entire dataset
emails['cleaned_text'] = emails['text'].apply(clean_text)

# Show sample results
print("Original vs Cleaned:")
for i in range(3):
    print(f"\n--- Email {i+1} ---")
    print(f"Original: {emails['text'].iloc[i][:150]}...")
    print(f"Cleaned:  {emails['cleaned_text'].iloc[i][:150]}...")

# Dataset statistics
print(f"\n\nDataset Shape: {emails.shape}")
print(f"Spam: {emails['spam'].sum()}, Ham: {(emails['spam']==0).sum()}")
```
