# Feature Engineering

This notebook covers feature extraction (TF-IDF, Bag-of-Words, and custom features) for spam classification.

```python
import pandas as pd
import yaml
from src.features import build_vectorizer, fit_transform, transform, save_vectorizer

# Load configuration
cfg = yaml.safe_load(open("../config.yaml"))
fc = cfg["features"]

# Load preprocessed data
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")

# Use cleaned text column if available, otherwise use original text
text_col = "cleaned_text" if "cleaned_text" in train.columns else "text"

X_train = train[text_col].fillna("")
X_test = test[text_col].fillna("")
y_train = train["spam"]
y_test = test["spam"]

# Build and fit vectorizer
vectorizer = build_vectorizer(
    method=fc["method"],
    max_features=fc["max_features"],
    ngram_range=tuple(fc["ngram_range"]),
    sublinear_tf=fc.get("sublinear_tf", True)
)

# Transform data
X_train_vec = fit_transform(vectorizer, X_train)
X_test_vec = transform(vectorizer, X_test)

print(f"Vectorizer: {fc['method'].upper()}")
print(f"Vocabulary size: {len(vectorizer.vocabulary_)}")
print(f"Train shape: {X_train_vec.shape}")
print(f"Test shape: {X_test_vec.shape}")

# Save vectorizer for later use
save_vectorizer(vectorizer, "../models/vectorizer.pkl")
```
