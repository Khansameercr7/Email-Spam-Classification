# Exploratory Data Analysis (EDA)

This notebook explores the email spam dataset, visualizes class distribution, and inspects sample emails.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
emails = pd.read_csv('../data/emails.csv')
emails.head()
```

```python
# Class distribution
sns.countplot(x='spam', data=emails, palette=['#3B8BD4', '#E8593C'])
plt.xticks([0, 1], ['Ham', 'Spam'])
plt.title('Class Distribution')
plt.show()
```

```python
# Show a few spam and ham emails
print('--- Example Ham Email ---')
display(emails[emails['spam'] == 0]['text'].iloc[0])
print('\n--- Example Spam Email ---')
display(emails[emails['spam'] == 1]['text'].iloc[0])
```

```python
# Dataset overview
print("Dataset Shape:", emails.shape)
print("\nColumn Types:")
print(emails.dtypes)
print("\nMissing Values:")
print(emails.isnull().sum())
print("\nBasic Statistics:")
emails.describe()
```

```python
# Text length analysis
emails['text_length'] = emails['text'].apply(len)
emails['word_count'] = emails['text'].apply(lambda x: len(str(x).split()))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Text length distribution
axes[0].hist(emails[emails['spam']==0]['text_length'], bins=50, alpha=0.7, label='Ham', color='#3B8BD4')
axes[0].hist(emails[emails['spam']==1]['text_length'], bins=50, alpha=0.7, label='Spam', color='#E8593C')
axes[0].set_xlabel('Text Length')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Email Text Length Distribution')
axes[0].legend()

# Word count distribution
axes[1].hist(emails[emails['spam']==0]['word_count'], bins=50, alpha=0.7, label='Ham', color='#3B8BD4')
axes[1].hist(emails[emails['spam']==1]['word_count'], bins=50, alpha=0.7, label='Spam', color='#E8593C')
axes[1].set_xlabel('Word Count')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Email Word Count Distribution')
axes[1].legend()

plt.tight_layout()
plt.show()

# Summary stats by class
print("\nText Length by Class:")
print(emails.groupby('spam')['text_length'].describe())
print("\nWord Count by Class:")
print(emails.groupby('spam')['word_count'].describe())
```
